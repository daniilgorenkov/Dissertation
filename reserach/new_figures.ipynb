```python
def set_new_columns(data:pd.DataFrame,N:str="Vertical"):
    new_cols = ["timestamp",N]
    data.columns = new_cols
    return data.set_index("timestamp")
```

```python
def plot_forces(data:pd.DataFrame,xlim:tuple=None,ylim:tuple=None):
    plt.figure().set_size_inches(12,8)
    sns.lineplot(set_new_columns(data))
    plt.grid(True)
    plt.xlabel(", ")
    plt.ylabel(" , ")
    plt.xlim(xlim)
    plt.ylim(ylim)
    plt.show()

plot_forces(normal,(1,4),(0,350000))
```
